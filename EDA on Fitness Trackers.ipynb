# Exploratory Data Analysis of Fitness Trackers available in the Indian Market

# Obtain and review raw data 

The Dataset has used in this analysis from Kaggle website. In this contempoarary world, the fitness tracker band is wearable device is simply a device that can be worn by the user and this device is capable of collecting vital information about the activities and health of the user through smart sensors.

In simple words, a fitness tracker can be worn by the user to collect and monitor data on fitness levels of the user through various activities like daily steps or total distance walked per day, hours slept with sleep stages, type of exercises with calories burnt, heartbeat monitoring, etc.

These devices are connected to the internet and are capable of syncing the recorded data to a cloud. The user can access this data in the form of a dashboard for their fitness monitoring requirements.

# Importing the important python libraries 

```python
#import pandas
import pandas as pd 
#import numpy
import numpy as np
#import seaborn
import seaborn as sns
#import matplotlib 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Fitness_excel_csv.csv" , thousands=',')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe()
```

```python
# Checking of missing values 
df.isna().sum().sort_values(ascending = False)
```

```python
# Replacing NAN values with 0
df = df.fillna(0)
df.iloc[:5]
```

```python
df.shape
```

```python
df.describe()
```

```python
df.dtypes
```

```python
sns.set_style("white")
sns.pairplot(df, hue='Brand Name')
```

# Count of products in the dataset 



```python
df['Brand Name'].groupby(df['Device Type']).count().sort_values(ascending = False)
```

# Count of Brands


```python
df['Brand Name'].nunique()
```

```python
df['Brand Name'].groupby(df['Brand Name']).count().sort_values(ascending = False)
```

# Average ratings by Brands 

```python
df.groupby('Brand Name')['Rating (Out of 5)'].mean().sort_values(ascending=False).iloc[:10]
```

# Rating vs Selling price

```python
fig, ax = plt.subplots(figsize=(15,3))
ax = sns.stripplot(x="Rating (Out of 5)", y="Selling Price", data=df)
```

# Brand wise mean Average Battery Life

```python
df.groupby('Brand Name')['Average Battery Life (in days)'].mean().sort_values(ascending=False).iloc[:10]
```

```python
sns.set_style("white")
sns.displot(df, x='Selling Price',bins=[5000,10000,20000,30000,40000,50000])
sns.displot(df, x='Original Price',bins=[5000,10000,20000,30000,40000,50000])
```

# Count of Strap Material

```python
df['Strap Material'].groupby(df['Strap Material']).count().sort_values(ascending=False).iloc[:10]
```

# Brandwise Mean Selling prices

```python
df.groupby('Brand Name')['Selling Price'].mean().sort_values(ascending=False).apply(np.ceil)
```

# Box plot to denote average selling prices


```python
fig, ax = plt.subplots(figsize=(22,5))
ax = sns.boxplot(x="Brand Name", y="Original Price", data=df)
```

```python
fig, ax = plt.subplots(figsize=(22,5))
ax = sns.stripplot(y="Selling Price", x="Brand Name", data=df)
```

```python
fig, ax = plt.subplots(figsize=(22,5))
ax = sns.stripplot(y="Original Price", x="Brand Name", data=df)
```
